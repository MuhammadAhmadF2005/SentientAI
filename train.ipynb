```python
import pandas as pd
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    'Name':    ['Ahmed', 'Sara', 'James', 'Lena', 'Omar', 'Priya', 'Tom', 'Nadia'],
    'Country': ['UAE', 'UK', 'USA', 'Germany', 'UAE', 'India', 'USA', 'UK'],
    'Age':     [28, 34, 45, 29, 52, 38, 41, 26],
    'Revenue': [1200, 850, 3400, 920, 150, 2100, 780, 1650],
    'Orders':  [5, 3, 12, 4, 1, 8, 3, 6]
})

df.index = [101, 102, 103, 104, 105, 106, 107, 108]
print(df)
```

```python
df.info()
```

```python
print(df.iloc[2])

print(df.loc[103])

print(df[['Revenue', 'Name']])

print(df[df['Revenue'] > 1000])


# Q6: UAE customers with more than 2 orders
uae_loyal_customers = df[(df['Country'] == 'UAE') & (df['Orders'] > 2)]
print(uae_loyal_customers)

# Q7: New column RevenuePerOrder
# Creating a new column by dividing Revenue by Orders
df['RevenuePerOrder'] = df['Revenue'] / df['Orders']
print(df[['Name', 'RevenuePerOrder']])

# Q8: Total revenue by country
total_revenue_by_country = df.groupby('Country')['Revenue'].sum()
print(total_revenue_by_country)

# Q9: Average age by country
average_age_by_country = df.groupby('Country')['Age'].mean()
print(average_age_by_country)

# Q10: Drop null Revenue rows (after setting James's revenue to None)
# Set James's revenue to None (NaN)
df.loc[df['Name'] == 'James', 'Revenue'] = np.nan

# Drop rows where Revenue is null
df_cleaned = df.dropna(subset=['Revenue'])
print(df_cleaned)


```
